```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import cv2
```

```python
img = plt.imread("diversity-people-group-team-union-260nw-379530769.jpg")
```

```python
plt.imshow(img)
```

```python
detector = cv2.CascadeClassifier(r"C:\Users\r8rey\AppData\Local\Programs\Python\Python38\Scripts\python\Face Detection\haarcascade_frontalface_alt.xml")
```

```python
detector
```

```python
faces = detector.detectMultiScale(img)
```

```python
for face in faces:
    x,y,w,h = face
    img = cv2.rectangle(img, (x,y), (x+w, y+h), (255,255,255), 2)
```

```python
plt.imshow(img)
```
